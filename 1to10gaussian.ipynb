```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu, sigma = 5.5, 1.4
sampleSize = 1000000
sample = np.random.normal(mu, sigma, sampleSize)
sample = np.round(sample)

count, bins, ignored = plt.hist(sample, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.show()
```

```python
bottomLine = 7
above = []
for x in sample:
    if x >= bottomLine:
        above.append(x)
```

```python
np.mean(above)
```
